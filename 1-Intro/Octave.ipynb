```octave
x = -0:0.01:3*pi; # Create an evenly-spaced vector from -10..10
y = x .* sin (3*x);    # y is also a vector
plot (x, y,"linewidth",5);
title ("Simple 2-D Plot");
grid();
xlabel ("x");
ylabel ("x sin(3x)");
```
